```python
import tkinter
from tkinter import *
a=Tk()
a.title("CALCULATOR")
v=("Arial",20)
a.geometry('470x500')
#CLEAR FUNCTION
def clr():
    e.delete(0,END)
#CLICKING VALUES IN ENTRY BOX
def clk(n):
    x=e.get()
    clr()
    av=x+n
    e.insert(0,av)
fn=0
sign=''
final=0
#BEFORE PRESS +,-,*,/ FUNCTIONS
def samd(n):
    global sign,fn
    sign=n
    fn=e.get()
    clr()
 #EQUAL FUNCTION   
def equal():
    global fn
    sn=e.get()
    clr()

    '''
    if sign=='+':
        final=int(fn)+int(sn)
    elif sign=='-':
        final=int(fn)-int(sn)
    elif sign=='x':
        final=int(fn)*int(sn)
    elif sign=='/':
        final=int(fn)/int(sn)
    elif sign=='%':
        final=int(fn)%int(sn)
    final=str(final)'''
    e.insert(0,eval(sn))
def back():
    c1=e.get()
    clr()
    if c1=='None':
        pass
    else :
        c2=int(c1)//10
        e.insert(0,str(c2))

#BUTTONS   
e=Entry(a,width=13,font=('Arial',45),justify=RIGHT)
e.grid(row=1,columnspan=4)
b1=Button(a,text='1',padx=35,pady=15,font=v,command=lambda:clk('1'))
b1.grid(row=5,column=0)
b2=Button(a,text='2',padx=35,pady=15,font=v,command=lambda:clk('2'))
b2.grid(row=5,column=1)
b3=Button(a,text='3',padx=35,pady=15,font=v,command=lambda:clk('3'))
b3.grid(row=5,column=2)
b4=Button(a,text='4',padx=35,pady=15,font=v,command=lambda:clk('4'))
b4.grid(row=4,column=0)
b5=Button(a,text='5',padx=35,pady=15,font=v,command=lambda:clk('5'))
b5.grid(row=4,column=1)
b6=Button(a,text='6',padx=35,pady=15,font=v,command=lambda:clk('6'))
b6.grid(row=4,column=2)
b7=Button(a,text='7',padx=35,pady=15,font=v,command=lambda:clk('7'))
b7.grid(row=3,column=0)
b8=Button(a,text='8',padx=35,pady=15,font=v,command=lambda:clk('8'))
b8.grid(row=3,column=1)
b9=Button(a,text='9',padx=35,pady=15,font=v,command=lambda:clk('9'))
b9.grid(row=3,column=2)
b0=Button(a,text='0',padx=35,pady=15,font=v,command=lambda:clk('0'))
b0.grid(row=6,column=0)
bp=Button(a,text='+',padx=33,pady=57,font=v,command=lambda:clk('+'))
bp.grid(row=5,column=3,rowspan=2)
bs=Button(a,text='-',padx=37,pady=15,font=v,command=lambda:clk('-'))
bs.grid(row=4,column=3)
bm=Button(a,text='x',padx=35,pady=15,font=v,command=lambda:clk('*'))
bm.grid(row=3,column=3)
bd=Button(a,text='/',padx=37,pady=15,font=v,command=lambda:clk('/'))
bd.grid(row=2,column=3)
bmod=Button(a,text='%',padx=30,pady=15,font=v,command=lambda:clk('%'))
bmod.grid(row=2,column=2)
be=Button(a,text='=',padx=35,pady=15,font=v,command=equal)
be.grid(row=6,column=2)
bdot=Button(a,text='.',padx=38,pady=15,font=v,command=lambda:clk('.'))
bdot.grid(row=6,column=1)
bb=Button(a,text='Back',padx=12,pady=15,font=v,command=back).grid(row=2,column=0)
bc=Button(a,text='Clear',padx=10,pady=15,font=v,command=clr)
bc.grid(row=2,column=1)
a.mainloop()
```
